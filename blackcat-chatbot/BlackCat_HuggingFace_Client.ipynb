```python
# Replace with your Hugging Face Space URL
# Format: https://<your-username>-<space-name>.hf.space
HF_SPACE_URL = "https://YOUR_USERNAME-blackcat-api.hf.space"

# Strip trailing slash if present
HF_SPACE_URL = HF_SPACE_URL.rstrip("/")

print(f"API URL : {HF_SPACE_URL}")
print(f"Docs   : {HF_SPACE_URL}/docs")
```

```python
import requests

def health_check():
    """Verify the Space is up before sending questions."""
    try:
        resp = requests.get(f"{HF_SPACE_URL}/health", timeout=30)
        resp.raise_for_status()
        print("API is healthy:", resp.json())
    except requests.exceptions.RequestException as e:
        print(f"Cannot reach API: {e}")
        print("   → Make sure your Space is running and the URL is correct.")

health_check()
```

```python
def ask_vuln(question: str):
    """
    Send a vulnerability question to the hosted CyberFixBot API
    on Hugging Face Spaces and print the structured answer.
    """
    url = f"{HF_SPACE_URL}/ask"
    payload = {"question": question}

    try:
        resp = requests.post(url, json=payload, timeout=60)
        resp.raise_for_status()
        data = resp.json()

        print(f"\n{'='*60}")
        print(f"Q: {data['question']}")
        print(f"{'='*60}")
        print(f"A:\n{data['answer']}")
        print(f"\nSources used: {data['sources_count']} chunks")
        return data

    except requests.exceptions.HTTPError as e:
        print(f"HTTP error {resp.status_code}: {resp.text}")
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")
```

```python
ask_vuln("How to fix XSS vulnerability?")
```

```python
ask_vuln("How to fix Critical SQL Injection?")
```

```python
# Interactive loop — type 'exit' to quit
print("🛡️ CyberFixBot — Type your vulnerability question (or 'exit' to quit)\n")
while True:
    user_input = input("You: ").strip()
    if user_input.lower() in ("exit", "quit", ""):
        print("Goodbye!")
        break
    ask_vuln(user_input)
```
